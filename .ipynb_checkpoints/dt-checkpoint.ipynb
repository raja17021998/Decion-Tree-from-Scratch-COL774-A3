```python
def oneHotEncode(data,to_one_hot_encode_cols):
    encoder = OneHotEncoder(sparse_output=False)
    one_hot_encoded = encoder.fit_transform(data[to_one_hot_encode_cols])
    encoded_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(to_one_hot_encode_cols))
    return encoded_df
```

```python
import numpy as np

def one_hot_encode(data):
    unique_values = np.unique(data)
    encoded_data = np.zeros((len(data), len(unique_values)), dtype=int)
    
    for i, value in enumerate(data):
        encoded_data[i, np.where(unique_values == value)] = 1
    
    return encoded_data

# Example usage:
data = np.array(['cat', 'dog', 'dog', 'bird', 'cat'])
encoded_data = one_hot_encode(data)
print(encoded_data)
```

```python
import pandas as pd
import numpy as np

class CustomOneHotEncoder:
    def __init__(self, column_name):
        self.column_name = column_name
        self.categories = None

    def fit_transform(self, input_df):
        if self.column_name not in input_df.columns:
            raise ValueError(f"Column '{self.column_name}' not found in the input DataFrame.")

        # Extract the column to be one-hot encoded
        data = input_df[self.column_name]
        self.categories = data.unique()

        one_hot_encoded_data = []
        for category in self.categories:
            one_hot_encoded_data.append(data.apply(lambda x: 1 if x == category else 0))

        one_hot_df = pd.DataFrame(np.array(one_hot_encoded_data).T, columns=self.categories)
        result_df = pd.concat([input_df, one_hot_df], axis=1)
        result_df = result_df.drop(columns=[self.column_name])

        return result_df

    def transform(self, input_df):
        if self.column_name not in input_df.columns:
            raise ValueError(f"Column '{self.column_name}' not found in the input DataFrame.")

        if self.categories is None:
            raise ValueError("Call 'fit_transform' before calling 'transform'.")

        # Extract the column to be one-hot encoded
        data = input_df[self.column_name]

        one_hot_encoded_data = []
        for category in self.categories:
            one_hot_encoded_data.append(data.apply(lambda x: 1 if x == category else 0))

        one_hot_df = pd.DataFrame(np.array(one_hot_encoded_data).T, columns=self.categories)
        result_df = pd.concat([input_df, one_hot_df], axis=1)
        result_df = result_df.drop(columns=[self.column_name])

        return result_df

# Example usage:
data = {'team': ['A', 'B', 'A', 'C', 'B']}
input_df = pd.DataFrame(data)

encoder = CustomOneHotEncoder(column_name='team')

# Fit and transform
result_df = encoder.fit_transform(input_df)
print("One-Hot Encoded DataFrame (fit_transform):")
print(result_df)

# Transform using the pre-fitted categories
result_df_transformed = encoder.transform(input_df)
print("\nOne-Hot Encoded DataFrame (transform):")
print(result_df_transformed)
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
import pandas as pd
import numpy as np

label_encoder = None 

def get_np_array(file_name):
    global label_encoder
    data = pd.read_csv(file_name)
    
    need_label_encoding = ['team','host','opp','month', 'day_match']
    if(label_encoder is None):
        label_encoder = OrdinalEncoder()
        label_encoder.fit(data[need_label_encoding])
    data_1 = pd.DataFrame(label_encoder.transform(data[need_label_encoding]), columns = label_encoder.get_feature_names_out())
    
    #merge the two dataframes
    dont_need_label_encoding =  ["year","toss","bat_first","format" ,"fow","score" ,"rpo" ,"result"]
    data_2 = data[dont_need_label_encoding]
    final_data = pd.concat([data_1, data_2], axis=1)
    
    X = final_data.iloc[:,:-1]
    y = final_data.iloc[:,-1:]
    return X.to_numpy(), y.to_numpy()
```

```python
import numpy as np

class TreeNode:
    def __init__(self, depth, max_depth):
        self.depth = depth
        self.max_depth = max_depth
        self.split_attribute = None
        self.split_value = None
        self.is_leaf = False
        self.class_label = None
        self.children = {}

def entropy(y):
    ones = np.count_nonzero(y == 1)
    zeros = np.count_nonzero(y == 0)
    total = ones + zeros

    if total == 0:
        return 0

    prob_0 = zeros / total
    prob_1 = ones / total

    if prob_0 in (0, 1) or prob_1 in (0, 1):
        return 0

    ent = -(prob_0 * np.log2(prob_0) + prob_1 * np.log2(prob_1))

    return ent

def information_gain(X, y, attribute, entropy_dataset, is_continuous):
    if is_continuous:
        idx0, idx1 = split_continuous_attribute(X, y, attribute)
        X_split_0, X_split_1, y_split_0, y_split_1 = X[idx0], X[idx1], y[idx0], y[idx1]
        info_gain = information_gain_continuous(X, y, attribute, entropy_dataset, idx0, idx1)
    else:
        indices_dict = split_categorical(X, y, attribute)
        info_gain = information_gain_categorical(X, y, attribute, entropy_dataset, indices_dict)

    return info_gain

def information_gain_continuous(X, y, attribute, entropy_dataset, idx0, idx1):
    prob_0 = len(idx0) / len(y)
    prob_1 = len(idx1) / len(y)
    entropy_attribute = prob_0 * entropy(y[idx0]) + prob_1 * entropy(y[idx1])
    info_gain = entropy_dataset - entropy_attribute
    return info_gain

def information_gain_categorical(X, y, attribute, entropy_dataset, indices_dict):
    entropy_attribute = 0
    total_values = len(y)

    for key in indices_dict:
        indices = indices_dict[key]
        prob_attr = len(indices) / total_values
        entropy_attribute += prob_attr * entropy(y[indices])

    info_gain = entropy_dataset - entropy_attribute
    return info_gain

def split_continuous_attribute(X, y, attribute):
    lis = X[:, attribute].astype(float)
    med = np.median(lis)
    idx0 = np.argwhere(lis <= med).flatten()
    idx1 = np.argwhere(lis > med).flatten()
    return idx0, idx1

def split_categorical(X, y, attribute):
    unique_values = np.unique(X[:, attribute])
    indices_dict = {}
    for unique_value in unique_values:
        indices = np.where(X[:, attribute] == unique_value)[0]
        indices_dict[unique_value] = indices
    return indices_dict

def build_decision_tree(X, y, max_depth, depth=0):
    unique_classes = np.unique(y)

    if depth == max_depth or len(unique_classes) == 1:
        node = TreeNode(depth, max_depth)
        node.is_leaf = True
        node.class_label = majority_vote(y)
        return node

    entropy_dataset = entropy(y)
    num_features = X.shape[1]
    best_info_gain = 0
    best_attribute = None
    best_split_value = None

    for attribute in range(num_features):
        is_continuous = attribute in [11]  # Change this list to specify continuous attributes
        info_gain = information_gain(X, y, attribute, entropy_dataset, is_continuous)

        if info_gain > best_info_gain:
            best_info_gain = info_gain
            best_attribute = attribute
            best_split_value = None if is_continuous else None

    if best_info_gain == 0:
        node = TreeNode(depth, max_depth)
        node.is_leaf = True
        node.class_label = majority_vote(y)
        return node

    node = TreeNode(depth, max_depth)
    node.split_attribute = best_attribute
    node.split_value = best_split_value
    node.children = {}

    if best_attribute in [11]:  # If it's a continuous attribute
        idx0, idx1 = split_continuous_attribute(X, y, best_attribute)
        node.children[0] = build_decision_tree(X[idx0], y[idx0], max_depth, depth + 1)
        node.children[1] = build_decision_tree(X[idx1], y[idx1], max_depth, depth + 1)
    else:
        indices_dict = split_categorical(X, y, best_attribute)
        for key in indices_dict:
            node.children[key] = build_decision_tree(X[indices_dict[key]], y[indices_dict[key]], max_depth, depth + 1)

    return node

def majority_vote(y):
    uniq_elements, counts = np.unique(y, return_counts=True)
    majority_element = uniq_elements[np.argmax(counts)]
    return majority_element

def predict(tree, X):
    predictions = []

    for sample in X:
        prediction = predict_single(tree, sample)
        predictions.append(prediction)

    return predictions

def predict_single(tree, sample):
    if tree.is_leaf:
        return tree.class_label

    if tree.split_attribute in [11]:  # Check if it's a continuous attribute
        if sample[tree.split_attribute] <= tree.split_value:
            return predict_single(tree.children[0], sample)
        else:
            return predict_single(tree.children[1], sample)
    else:
        attribute_value = sample[tree.split_attribute]
        if attribute_value in tree.children:
            return predict_single(tree.children[attribute_value], sample)
        else:
            return majority_vote(sample)

max_depth = 10
X_train,y_train = get_np_array('data/train.csv')
X_test, y_test = get_np_array("data/test.csv")
decision_tree = build_decision_tree(X_train, y_train, max_depth)
```
